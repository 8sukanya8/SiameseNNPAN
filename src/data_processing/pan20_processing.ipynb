```python
import os
import fnmatch
from random import shuffle
import json
from src.config import config_io
from pathlib import Path
import json
import pandas as pd
```

```python
data_path = config_io['pan_20_original_test_narrow']
#data_path = config_io['pan_20_original_train_narrow']
print(data_path)
base = [file.split(".")[0] for file in os.listdir(data_path) if "txt" in file]
problem_files = [file+".txt" for file in base]
truth_files = ["truth-"+file+".json" for file in base]
```

```python
def verify_file_exists(file_list,data_path):
    for file in file_list:
        if not Path(data_path+file).exists():
            print("Error! Could not find file: ",data_path+file)
```

```python
verify_file_exists(truth_files,data_path)
verify_file_exists(problem_files,data_path)
```

Split text into paragraphs.Make author list according to structure and changes. Then assign paragraphs to corresponding authors. Fol

```python
a1_list = []
a2_list = []
p1_list = []
p2_list = []
p_list = []

for i in range(0,4):#len(truth_files)):
    truth_file = truth_files[i]
    problem_file = problem_files[i]
    print((truth_file, problem_file))
    with open(data_path+problem_file) as f:
        text = f.read()
    print("len:", len(text))
    with open(data_path+truth_file) as json_file:
        data = json.load(json_file)
    structure = data['structure']
    changes = data['changes']
    count = 0
    authors = [structure[count]]
    for i in range(0,len(changes)):
        if changes[i]==0:
            authors.append(structure[count])
        else:
            count+=1
            authors.append(structure[count])
    paragraphs = text.split("\n\n")
    a1 = [authors[i-1] for i,j in enumerate(authors) if i>=1]
    a2 = [authors[i] for i,j in enumerate(authors) if i>=1]
    p1 = [paragraphs[i-1] for i,j in enumerate(paragraphs) if i>=1]
    p2 = [paragraphs[i] for i,j in enumerate(paragraphs) if i>=1]
    for author in a1:
        a1_list.append(author)
    for author in a2:
        a2_list.append(author)
    for para in p1:
        p1_list.append(para)
    for para in p2:
        p2_list.append(para)
    for author in a1: #just adding the problem files as many times as rows are present
        p_list.append(problem_file)
data = {"problem":p_list, "author_1":a1_list, "author_2":a2_list, "para1_text": p1_list, "para2_text":p2_list}
df = pd.DataFrame(data)
```
